### loc을 기준으로 데이터 구조 변경

```python
def locs(x):
    if int(x[-1]) == 1:
        return 1
    elif int(x[-1]) == 2:
        return 2
    else: return 3
```

### 24시간 후

```python
def after_24H(x):
    date = pd.to_datetime(x)
    hour_after_24 =  date + datetime.timedelta(days=1)
    return str(hour_after_24)
```

```python
data['plant_train.x24h_cond_tma'] = data['plant_train.mea_ddhr'].apply(lambda x: after_24H(x)).astype(object)
```

```python
data = data.rename(columns={'plant_train.x24h_cond_tma' : 'merging_key'})
```

#### 24시간후 condition 추가

```python
after24 = data[['plant_train.mea_ddhr', 'plant_train.plant','plant_train.loc','plant_train.cond_loc']]

after24 = after24.rename(columns={'plant_train.cond_loc':'plant_train.x24h_cond_loc',
                                 'plant_train.mea_ddhr' : 'merging_key'})

after24['merging_key'] = after24['merging_key'].astype(str)

after24 = after24.sort_values(by='merging_key').reset_index().drop(columns='index', axis=1)

a = data.merge(after24, 
                on=['merging_key','plant_train.plant','plant_train.loc'], 
                how = 'left').sort_values(by='plant_train.mea_ddhr').reset_index().drop(columns='index',axis=1)
```

### 48시간 후

```python
def after_48H(x):
    date = pd.to_datetime(x)
    hour_after_48 =  date + datetime.timedelta(days=2)
    return str(hour_after_48)
```

```python
data['merging_key'] = data['plant_train.mea_ddhr'].apply(lambda x: after_48H(x))
```

#### 48시간후 condition 추가

```python
after48 = data[['plant_train.mea_ddhr', 'plant_train.plant','plant_train.loc','plant_train.cond_loc']]

after48 = after48.rename(columns={'plant_train.cond_loc':'plant_train.x48h_cond_loc',
                                 'plant_train.mea_ddhr' : 'merging_key'})

after48['merging_key'] = after48['merging_key'].astype(str)

after48 = after48.sort_values(by='merging_key').reset_index().drop(columns='index', axis=1)

b = data.merge(after48, 
                on=['merging_key','plant_train.plant','plant_train.loc'], 
                how = 'left').sort_values(by='plant_train.mea_ddhr').reset_index().drop(columns='index',axis=1)
b.shape
```
