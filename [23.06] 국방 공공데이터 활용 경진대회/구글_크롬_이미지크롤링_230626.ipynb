```python
from selenium import webdriver
import time
import urllib.request
import os

# 이미지를 저장할 디렉토리 설정
root = "images"

# 검색어 설정
search_query = "gun"

# chromedriver 경로
chromedriver_path = os.path.join(os.getcwd(), "chromedriver.exe")  # 64비트는 Linux64bit 버전을 다운받는다.

# 웹 드라이버 실행
driver = webdriver.Chrome(chromedriver_path)

# 구글 이미지 검색 페이지 열기
driver.get("https://www.google.com/imghp")

# 검색어 입력
search_box = driver.find_element_by_name("q")
search_box.send_keys(search_query)
search_box.submit()

# 이미지 링크 가져오기
image_links = []
for i in range(5):  # 페이지 스크롤 횟수 조정
    driver.execute_script("window.scrollBy(0,10000)")
    time.sleep(2)
    images = driver.find_elements_by_css_selector(".rg_i")
    for image in images:
        link = image.get_attribute("src")
        if link and link.startswith("http"):
            image_links.append(link)

# 이미지 다운로드
for i, link in enumerate(image_links[:100], 1):  # 다운로드할 이미지 수 조정
    try:
        file_name = f"{root}/{search_query}_{i}.jpg"
        urllib.request.urlretrieve(link, file_name)
        print(f"Downloaded image {i}")
    except:
        print(f"Failed to download image {i}")

# 웹 드라이버 종료
driver.quit()
```
