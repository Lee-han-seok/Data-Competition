```python
import csv 
```

```python
# CSV파일 텍스트를 읽어오는 작업
# csv 파일은 1열로 구성된 구글폼 답안임.
def extract_words_from_csv(file_path):
    words = []

    with open(file_path, 'r') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            for word in row:
                words.append(word)

    return words

# CSV 파일 경로 지정
csv_file_path = '청춘앨범_만족이유(받게된다면).csv'

# 단어 추출
word_list = extract_words_from_csv(csv_file_path)

# 추출된 단어 출력
for word in word_list:
    print(word)
```

```python
def remove_punctuation(word_list):
    new_word_list = []

    for word in word_list:
        new_word = word.replace(",", "").replace("'", "")
        new_word_list.append(new_word)

    return new_word_list

# 단어 리스트에서 쉼표와 작은따옴표 제거
updated_word_list = remove_punctuation(word_list)

# 쉼표와 작은따옴표가 제거된 단어 출력
for word in updated_word_list:
    print(word)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def extract_nouns(word_list):
    nouns = []

    for word in word_list:
        tokens = word_tokenize(word)
        tagged_words = pos_tag(tokens)

        for tagged_word in tagged_words:
            if tagged_word[1].startswith('NN'):  # 명사 품사 태그 확인
                nouns.append(tagged_word[0])

    return nouns

# 단어 리스트에서 명사 추출
noun_list = extract_nouns(word_list)

# 추출된 명사 출력
for noun in noun_list:
    print(noun)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import os

def generate_wordcloud(word_list):
    # 공백으로 단어들을 연결하여 하나의 문자열로 생성
    text = ' '.join(word_list)

    # 현재 작업 디렉토리의 경로를 가져옴
    current_dir = os.getcwd()

    # 폰트 파일 경로 설정
    font_path = os.path.join(current_dir, 'NanumGothic.ttf') # 한글이 깨져서 현 디렉토리에 나눔고딕 다운로드 후 진행하였음.

    # 워드 클라우드 생성
    wordcloud = WordCloud(width=800, height=400, max_words=50, background_color='white', font_path=font_path).generate(text)

    # 워드 클라우드 이미지 표시
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# 단어 리스트로 워드 클라우드 생성
generate_wordcloud(word_list)
```
